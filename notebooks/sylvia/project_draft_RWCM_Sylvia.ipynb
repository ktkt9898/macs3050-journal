## Your new shiny data visualization

- **Explain what you show, might require math:** Since industries produce for two reasons (Reason 1: output becomes input for other industries which is called intermediate demand ; Reason 2: output goes directly to consumer so governments, people, and others which is called final demand), Total Output = Intermediate Demand + Final Demand.

- This equation can be rewritten as x = Ax + d where x is a column vector that tells you how much each industry produces in total, A is the input-output matrix that shows how much of each industry is needed to produce one unit of a specific industry so when you mutiply the two you get Ax which shows the actual total amount of inputs consumed in order to achieve the amount being produced. If you add up all of the values in the column vector Ax which shows the intermediate demand, then you get the total intermediate demand in the economy. It is important to consider that not all of the outputs, as mentioned before, become inputs and that some are final demand which is what d represents in the equation.

- Now if we wanted to solve the equation we would need to rearrange the variables so x = Ax + d which is also the exact same as I * x = Ax + d (where I is the identity matrix) can have Ax subtracted from both sides which in turn gives us (I-A)x = d so now we can solve for X in order to find out what the intermediate demand is for each industry which can then be solved using the Gaussian Elimination method.
  
- **Show:** The project will show how to find the total output and the intermediate demand by choosing a final demand as well as knowing what values are in the A matrix. Essentially the CSV file gives us the A matrix (how the industries depend on each other) and we can choose whatever final demand we want so we can then calculate the total output as well as the intermediate demand. We can also use a bar chart to show that total output is always bigger than final demand.
  
- **Comment:** Find which industries produce most relative to demand and gain a better understanding of the output - input model in practice.
- **Hypothesis:** We see that some industries (like Energy) amplify production needs across the system because many others rely on them.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV (skip metadata rows if needed)
df = pd.read_csv("Leontief_Matrix.csv", skiprows=3)

# Show the first few rows to confirm structure
display(df.head())

# Example bar chart:
# Let's take the first industry row and plot its requirements across all industries
row = df.iloc[0, 2:]   # numeric values (skip first 2 cols: industry code & description)

plt.figure(figsize=(14,6))
plt.bar(df.columns[2:], row)
plt.xticks(rotation=90)
plt.ylabel("Total Requirements")
plt.title(f"Industry Requirements for {df.iloc[0,1]} ({df.iloc[0,0]})")
plt.tight_layout()
plt.show()
```

```python
# | label: fig-label-to-make-reference
# | fig-cap: "Heatmap of Industries in the CSV file"
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV (skip metadata rows)
df = pd.read_csv("Leontief_Matrix.csv", skiprows=3)

# Drop the first two identifier columns: industry code + description
matrix = df.drop(columns=[df.columns[0], df.columns[1]])

# Make sure everything is numeric
matrix = matrix.apply(pd.to_numeric, errors="coerce")

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(matrix, cmap="Blues", cbar=True)
plt.title("Leontief Total Requirements Matrix", fontsize=14)
plt.xlabel("Industries (Outputs)")
plt.ylabel("Industries (Inputs)")
plt.show()
```
